```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Data Extraction: Loading and Initial Inspection of `raw_dataset.csv`

```python
file_path = '/content/raw_dataset.csv'
df = pd.read_csv(file_path, sep=';')
print(f"Successfully loaded data from {file_path}.\n")
```

```python
display(df.info())
```

#### Descriptive Statistics of numerical columns

```python
display(df.describe())
display(df.describe(include='all'))
```

#### Check Missing Values

```python
missing_values = df.isnull().sum()
print("Number of missing values per column:")
display(missing_values[missing_values > 0].sort_values(ascending=False))
```

#### Check Duplicates

```python
print(f"Number of duplicate rows: {df.duplicated().sum()}")
```

#### Unique Values (Column-wise)

```python
print("Number of unique values per column:")
display(df.nunique().sort_values(ascending=False))
```

#### Unique Value Types (Important for EDA later)

```python
for col in df.select_dtypes(include=['object', 'category']).columns:
    if df[col].nunique() < 50: # Display unique values for columns with less than 50 unique entries
        print(f"\nUnique values for column '{col}':")
        print(df[col].unique())
    else:
        print(f"\nColumn '{col}' has {df[col].nunique()} unique values (too many to display all). Top 10 most frequent values:")
        display(df[col].value_counts().head(10))
```

#### Basic Data Quality Observations

Based on the initial inspection:
- **Missing Values**: There are no missing values in any of the columns, as indicated by `df.isnull().sum()` returning all zeros.
- **Duplicate Rows**: There are 12 duplicate rows present in the dataset.
- **Data Types**: The data types (`int64`, `float64`, `object`) appear appropriate for the content of each column.
- **Unexpected Categories**: Some categorical columns like `job`, `marital`, `education`, `default`, `housing`, and `loan` contain an 'unknown' category. This will need to be addressed during data cleaning or feature engineering, as it represents missing or unclassified information.
- **Separators**: The dataset was loaded successfully using a semicolon (`;`) as a separator, which was crucial for correct parsing.
