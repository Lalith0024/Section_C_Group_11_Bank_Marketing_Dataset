### Importing Required Libraries

In this step, we import essential Python libraries:
- pandas → for data manipulation and analysis
- numpy → for numerical operations

These libraries will be used throughout the data cleaning process.

```python
import pandas as pd
import numpy as np
```

### Loading Raw Dataset

The dataset is loaded using pandas.

- The separator is set to `;` because the dataset is semicolon-separated.
- We check the shape to understand the size of the dataset.
- We display the first few rows to get an initial view of the data.

This helps us verify that the dataset is loaded correctly.

```python
df = pd.read_csv('/content/raw_dataset.csv', sep=';')
print("Shape:", df.shape)
df.head()
```

###  Basic Data Inspection

In this step, we explore the structure of the dataset:

- df.info() → shows column names, data types, and missing values
- df.describe() → provides statistical summary of numerical columns
- df.nunique() → shows the number of unique values in each column

This helps us understand data types, distributions, and potential issues.

```python
df.info()
```

```python
df.describe()
```

```python
df.nunique()
```

### Removing Duplicate Records

Duplicate rows can bias analysis and lead to incorrect insights.

- We remove duplicate entries using drop_duplicates()
- We verify the new dataset shape after removal

This ensures data integrity and avoids double counting.

```python
df = df.drop_duplicates()
print("After removing duplicates:", df.shape)
```

### Fixing Data Types

The 'nr_employed' column is converted to integer type.

- This ensures consistency in numerical operations
- Prevents issues during analysis and visualization

Proper data types are important for accurate computations.

```python
df['nr.employed'] = df['nr.employed'].astype(int)
```

### Renaming Columns for Better Clarity

Column names are standardized to improve readability and usability.

- Converted to descriptive and consistent naming format
- Replaced short or ambiguous names with meaningful ones

This makes the dataset easier to understand and work with.

```python
df.columns = [
    "age",
    "job_type",
    "marital_status",
    "education_level",
    "credit_default",
    "has_housing_loan",
    "has_personal_loan",
    "contact_type",
    "contact_month",
    "contact_day",
    "call_duration_sec",
    "num_contacts_campaign",
    "days_since_last_contact",
    "num_previous_contacts",
    "previous_campaign_outcome",
    "employment_variation_rate",
    "consumer_price_index",
    "consumer_confidence_index",
    "interest_rate",
    "number_of_employees",
    "subscribed"
]

df.head()
```

### Standardizing Values

In this step:

- The target column is converted from 'yes/no' to 1/0 format
- 'unknown' values are standardized to 'Unknown' for consistency

This ensures uniform data representation across the dataset.

```python
df["subscribed"] = df["subscribed"].map({"yes": 1, "no": 0})

df.replace("unknown", "Unknown", inplace=True)

df.head()
```

### Cleaning Education Levels

The 'illiterate' category is merged into 'basic.4y':

- Very few records exist for 'illiterate'
- Keeping it separate would create noise in analysis

This improves category consistency and reliability.

```python
df["education_level"] = df["education_level"].replace("illiterate", "basic.4y")
```

### Converting Columns to Categorical Type

Categorical columns are explicitly converted to 'category' datatype:

- Improves memory efficiency
- Helps visualization tools (like Tableau) interpret data correctly

This step prepares the dataset for analysis and dashboarding.

```python
categorical_cols = [
    "job_type", "marital_status", "education_level",
    "has_housing_loan", "has_personal_loan",
    "contact_type", "contact_month", "contact_day",
    "previous_campaign_outcome"
]

for col in categorical_cols:
    df[col] = df[col].astype("category")

df.info()
```

### Handling Outliers in Campaign Calls

The 'num_contacts_campaign' column contains extreme values (very high number of calls).

- We cap the maximum value at 10 using `.clip()`
- This prevents outliers from distorting analysis and visualizations

This ensures more realistic and stable insights.

```python
df["num_contacts_campaign"] = df["num_contacts_campaign"].clip(upper=10)

df["num_contacts_campaign"].describe()
```

### Creating Age Groups

The 'age' column is grouped into meaningful categories:

- 0–25 → Young
- 26–35 → Early Career
- 36–50 → Mid Career
- 51–65 → Senior
- 65+ → Retired

Labels are assigned in numeric order (1–5) to ensure correct sorting in Tableau.

This simplifies analysis and improves visualization.

```python
df["age_group"] = pd.cut(
    df["age"],
    bins=[0, 25, 35, 50, 65, 100],
    labels=[
        "1",
        "2",
        "3",
        "4",
        "5"
    ]
)
```

### 📞 Creating Call Intensity Feature

The number of campaign calls is grouped into buckets:

- 1 → First contact
- 2 → Second contact
- 3-5 → Moderate contact
- 6+ → High contact

This helps analyze how repeated calls affect customer conversion.

It is more interpretable than raw call counts.

```python
df["call_intensity"] = df["num_contacts_campaign"].apply(
    lambda x: "1" if x == 1 else
              "2" if x == 2 else
              "4" if 3 <= x <= 5 else
              "6"
)
```

### Identifying Previously Contacted Customers

A new column 'contacted_before' is created:

- Yes → if the customer was contacted previously
- No → if no previous contact

This helps analyze whether prior interaction influences current success.

```python
df["contacted_before"] = df["num_previous_contacts"].apply(
    lambda x: "Yes" if x > 0 else "No"
)
df[["num_previous_contacts", "contacted_before"]].head()
```

### Creating Contact Recency Feature

The 'days_since_last_contact' column is transformed into categories:

- 999 → Never Contacted
- ≤ 7 → Recently Contacted
- ≤ 30 → Moderately Contacted
- .> 30 → Long Ago Contacted

This converts a misleading numeric field into meaningful business categories.

It improves interpretability in analysis.

```python
df["contact_recency"] = df["days_since_last_contact"].apply(
    lambda x: "Never Contacted" if x == 999 else
              "Recently Contacted" if x <= 7 else
              "Moderately Contacted" if x <= 30 else
              "Long Ago Contacted"
)

df[["days_since_last_contact", "contact_recency"]].head()
```

### Removing Raw Recency Column

The original 'days_since_last_contact' column is dropped:

- Contains placeholder value (999) which distorts analysis
- Already replaced by 'contact_recency'

This avoids duplication and misleading interpretations.

```python
df.drop(columns=["days_since_last_contact"], inplace=True)
```

### Creating Final Target Variable

The 'subscribed' column is renamed to 'subscription_flag':

- Converted to binary format (1 = Yes, 0 = No)
- Old column is removed to avoid duplication

This becomes the main target variable for analysis.

```python
df["subscription_flag"] = df["subscribed"]
df.drop(columns=["subscribed"], inplace=True)
```

### Creating Engineered Features

Additional features are created to simplify analysis:

- rate_environment → categorizes interest rates into Low, Medium, High
- prev_success_flag → indicates if previous campaign was successful

These features improve interpretability and support better insights.

```python
# ADVANCED INSIGHT FEATURES

# 1. Rate Environment (simplifies economic conditions)
df['rate_environment'] = pd.cut(
    df['interest_rate'],
    bins=[0, 1.5, 3.5, 6],
    labels=['Low', 'Medium', 'High']
)

# w. Previous Success Flag (strongest predictor)
df['prev_success_flag'] = (df['previous_campaign_outcome'] == 'success').astype(int)

df[['rate_environment', 'prev_success_flag']].head()
```

### Creating Education Order for Sorting

An ordinal mapping is created for education levels:

- Assigns numerical values based on education hierarchy
- Ensures correct sorting in Tableau visualizations

This avoids incorrect alphabetical ordering.

```python
#  EDUCATION ORDER FIX (for Tableau sorting)

edu_map = {
    'basic.4y':1,
    'basic.6y':2,
    'basic.9y':3,
    'high.school':4,
    'professional.course':5,
    'university.degree':6,
    'Unknown':0
}

df['education_ordinal'] = df['education_level'].map(edu_map)
```

### Final Dataset Verification

Final checks are performed:

- Confirm dataset shape
- Display sample rows

This ensures all cleaning steps were applied correctly.

```python
print(df.shape)
df.head()
```

### Exporting Cleaned Dataset

The cleaned dataset is saved as a CSV file:

- Ready for EDA and dashboard creation
- Ensures reproducibility of results

This marks the completion of the data cleaning process.

```python
df.to_csv("cleaned_bank_data.csv", index=False)
print("✅ FINAL CLEANED DATASET READY")
```
